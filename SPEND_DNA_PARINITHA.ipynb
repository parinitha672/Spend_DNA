```python
#IMPORTING LINRARIES
import pandas as pd
import numpy as np
```

```python
#loading the dataset
df=pd.read_csv("Data set for DADS June.csv")
#display first five records
df.head()
```

```python
#check duplicate values
duplicate_records=df.duplicated().sum()
print("Duplicated Records :", duplicate_records)
```

```python
#remove duplicated records
df=df.drop_duplicates()
print("Remaining Records :",len(df))
```

```python
# Convert Date column into datetime format
df["Date"] = pd.to_datetime(
    df["Date"],
    dayfirst=True,
    errors="coerce"
)
df["Date"].head()
```

```python
# Clean Amount column
df["Amount"] = df["Amount"].astype(str)
df["Amount"] = df["Amount"].str.replace("₹", "")
df["Amount"] = df["Amount"].str.replace("Rs.", "")
df["Amount"] = df["Amount"].str.replace(",", "")
df["Amount"] = df["Amount"].str.strip()
df["Amount"] = pd.to_numeric(
    df["Amount"],
    errors="coerce"
)
df["Amount"].head()
```

```python
# Standardize transaction type
df["Type"] = df["Type"].str.lower()
df["Type"] = df["Type"].replace("dr", "debit")
df["Type"] = df["Type"].replace("cr", "credit")
df["Type"].value_counts()
```

```python
# Check missing values after cleaning
print(df.isnull().sum())
```

```python
# Remove rows with invalid dates or amounts
df = df.dropna(subset=["Date", "Amount"])
print("Clean Dataset Shape :", df.shape)
```

```python
# Function to identify vendors
def find_vendor(text):
    text = str(text).upper()
    for vendor in vendor_dict:
        for word in vendor_dict[vendor]:
            if word in text:
                return vendor
    return "Uncategorised"
```

```python
df["Vendor"] = df["Description"].apply(find_vendor)
uncategorised = df[df["Vendor"] == "Uncategorised"]
print("Remaining Uncategorised :", len(uncategorised))
uncategorised["Description"].drop_duplicates()
```

```python
# Apply vendor extraction
df["Vendor"] = df["Description"].apply(find_vendor)
df[["Description", "Vendor"]].head(15)
```

```python
# Calculate total credits and total debits
total_credit = credit_data["Amount"].sum()
total_debit = debit_data["Amount"].sum()
print("Total Credit :", total_credit)
print("Total Debit :", total_debit)
```

```python
# Extract month from Date column
df["Month"] = df["Date"].dt.month
df[["Date", "Month"]].head()
```

```python
# Display transaction count for each month
monthly_transactions = df["Month"].value_counts().sort_index()
monthly_transactions
```

```python
# Update debit and credit data after adding Month column
credit_data = df[df["Type"] == "credit"]
debit_data = df[df["Type"] == "debit"]
```

```python
#calculate monthly spending
monthly_spending = (
    debit_data.copy()
    .assign(Month=debit_data["Date"].dt.month)
    .groupby("Month")["Amount"]
    .sum()
    .sort_index()
)
monthly_spending
```

```python
# Create category-wise monthly spending table
monthly_category = pd.pivot_table(
    debit_data.copy().assign(Month=debit_data["Date"].dt.month),
    values="Amount",
    index="Category",
    columns="Month",
    aggfunc="sum",
    fill_value=0
)
monthly_category
```

```python
# Display category-wise monthly spending
print(monthly_category)
```

```python
# Calculate increase in spending from first month to last month
growth = {}
for category in monthly_category.index:
    first_month = monthly_category.loc[category].iloc[0]
    last_month = monthly_category.loc[category].iloc[-1]
    difference = last_month - first_month
    growth[category] = difference
growth
```

```python
# Find category with highest increase
highest_growth = max(growth, key=growth.get)
print("Highest Increase in Spending")
print(highest_growth)
print(growth[highest_growth])
```

```python
# Find category with highest decrease
lowest_growth = min(growth, key=growth.get)
print("Highest Decrease in Spending")
print(lowest_growth)
print(growth[lowest_growth])
```

```python
# Calculate category-wise average amount
category_mean = debit_data.groupby("Category")["Amount"].transform("mean")
category_mean.head()
```

```python
# Calculate category-wise standard deviation
category_std = debit_data.groupby("Category")["Amount"].transform("std")
category_std.head()
```

```python
# Calculate Z-Score
debit_data["Z_Score"] = (
    debit_data["Amount"] - category_mean
) / category_std
debit_data[["Amount", "Category", "Z_Score"]].head()
```

```python
# Identify unusual transactions
anomalies = debit_data[debit_data["Z_Score"] > 3]
print("Number of Anomalies :", len(anomalies))
```

```python
# Display anomaly transactions
anomalies[
    [
        "Date",
        "Description",
        "Vendor",
        "Category",
        "Amount",
        "Z_Score"
    ]
]
```

```python
# Sort anomalies by Z-Score
anomalies = anomalies.sort_values(
    by="Z_Score",
    ascending=False
)
anomalies.head(10)
```

```python
# Display top anomaly
if len(anomalies) > 0:
    print("Highest Anomaly")
    print(anomalies.iloc[0])
else:
    print("No anomalies found.")
```

```python
# Summary of anomalies
print("=" * 50)
print("ANOMALY DETECTION")
print("=" * 50)
print("Total Anomalies :", len(anomalies))
if len(anomalies) > 0:
    print()
    print(anomalies[
        [
            "Vendor",
            "Category",
            "Amount",
            "Z_Score"
        ]
    ].head())
```

```python
# Identify Late-Night Snacker
if late_food_percentage >= 50:
    archetypes.append("THE LATE-NIGHT SNACKER")
```
